```python
### Roger Valdez
### ISQS 6337: Scripting Languages
### Data Analysis Project Due: 10-11-2018
### Weather Data from World War II: 1940-1945
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
### Weather Temperature Data
weatherdata=pd.read_csv('Summary of Weather.csv')
weatherdata.head(10)
```

```python
### Drop data
weatherdata = weatherdata.drop(['Precip','Snowfall','STA','Date','WindGustSpd','PoorWeather',"PRCP",'DR','SPD','MaxTemp', 'MinTemp','MeanTemp', 'SNF', 'SND', 'FT', 'FB', 'FTI', 'ITH', 'PGT', 'TSHDSBRSGF', 'SD3', 'RHX', 'RHN', 'RVG', 'WTE'], axis=1)
weatherdata.head(30)
```

```python
### What were the weather conditions during World War II?
```

```python
### Use qqplot style
style.use('ggplot')

### Scatter plot for MaxTemp and MinTemp fitting a regression line
sns.regplot('MAX','MIN', data=weatherdata).set_title('Max and Min Temperature')
```

```python
### Scatterplot with varying point sizes
sns.relplot(x='MO',y='MIN',size='MAX',data=weatherdata)
plt.show()
```

```python
### Scatter plot 
fig, ax = plt.subplots(1,2,figsize=(12,6))
sns.boxplot(x='YR',y='MEA', data=weatherdata,ax=ax[0])
sns.boxplot(x='MO',y='MEA', data=weatherdata,ax=ax[1])
plt.show()
```
